## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
hist_data = pd.read_csv('https://raw.githubusercontent.com/Kiyosk-fc/Basic-Python/master/colab/quizzes/matplotlib_quiz/hist_data.csv')
hist_data['aデータ'].plot(kind="hist", alpha = 0.5, bins = 30, color = "orange")
hist_data['bデータ'].plot(kind="hist", alpha = 0.5, bins = 30, color = "c")
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend(loc = "upper right")
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
scatter_data = pd.read_csv("https://raw.githubusercontent.com/Kiyosk-fc/Basic-Python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv")
class0 = scatter_data[scatter_data["class"] == 0]
class1 = scatter_data[scatter_data["class"] == 1]
class2 = scatter_data[scatter_data["class"] == 2]
plt.scatter(class0["length (cm)"],class0["width (cm)"], color = "r", label = "dass0") 
plt.scatter(class1["length (cm)"],class1["width (cm)"], color = "b", label = "dass1") 
plt.scatter(class2["length (cm)"],class2["width (cm)"], color = "g", label = "dass2")
plt.grid()
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.legend(loc = "upper right",)
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
line_data = pd.read_csv("https://raw.githubusercontent.com/Kiyosk-fc/Basic-Python/master/colab/quizzes/matplotlib_quiz/line_data.csv")
plt.figure(figsize = [6.4,10])
plt.subplot(2,1,1)
plt.plot(line_data["期間"],line_data["件数"], marker ="o")
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.xticks(rotation = 30)
plt.ylabel("件数")
plt.grid()

plt.subplot(2,1,2)
plt.subplots_adjust(hspace = 1)
plt.bar(line_data["期間"],line_data["費用"])
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.xticks(rotation = 30)
plt.ylabel("費用(円)")
plt.grid()
```
